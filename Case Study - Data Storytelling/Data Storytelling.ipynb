## # of Cars between 1973 and 2020

Ride share programs have made it easier to get around, so it is important to see if people are still buying cars. 

```python
fig, ax = plt.subplots()

cars_by_year = df.groupby('year')['year'].count()

cars_by_year.plot(kind = 'line', figsize=(12, 5))

ax.set_xlabel('Year of Car Release', labelpad=15)
ax.set_ylabel('# of cars', labelpad=15)
ax.set_title('# of cars produced each year [1990 - 2020]', pad=15, weight='bold')

fig.autofmt_xdate()

fig.tight_layout()
```

## Mileage

I want to buy a car that will last a long time, looking at the average mileage for each manufacturer is a good indicator. However, there are newer manufacturers like Tesla, which will cause their average to be lower.

```python
fig, ax = plt.subplots()

mileage_colors = list('ggggggggggggggyyyyyyyyyyyyyyrrrrrrrrrrrrrr')

mileagemean = df.groupby('manufacturer')['odometer'].mean().sort_values(ascending = False)

mileagemean.plot(kind='barh', figsize=(10,7), color = mileage_colors)


ax.set_xlabel('Car Manufacturer', labelpad=15 )
ax.set_ylabel('Average Mileage', labelpad=15 )
ax.set_title('Average Car Mileage for Each Manufacturer', pad=15 ,
             weight='bold')

fig.tight_layout()
```

## Car Price

This is only looking at a few manufacturers that I would consider buying from. The graph below shows the range of car prices for each manufacturer over the past 5 years. 

```python
fig, ax = plt.subplots(figsize = (20, 12))



options = ['honda','toyota', 'nissan', 'mazda', 'volkswagen']
rslt_df = df[df['manufacturer'].isin(options)] 


data = rslt_df.loc[rslt_df['year'].between(2015, 2019)]
data_price = data.loc[data['price'] <= 50000]

sns.lineplot(data=data_price, x="year", y="price", hue = 'manufacturer')


plt.show()
```
